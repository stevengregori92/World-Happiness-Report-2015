#Data Visualization

# Report

Insights:
- The lower Standard Error, the higher Happiness Score
- The higher Economy, the higher Happiness Score
- The higher Family, the higher Happiness Score
- The higher Health, the higher Happiness Score
- The higher Freedom, the higher Happiness Score
- The lower Trust, the higher Happiness Score
- The lower Generosity, the higher Happiness Score
- The higher Dystopia Residual, the higher Happiness Score

Conclusion: Happiness Score in countries is affected by Standard Error, Economy, family, health, Freedom, Trust, Generosity, and Dystopia Residual.
